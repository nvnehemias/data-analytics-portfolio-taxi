```python
df = pd.read_parquet("../data/raw/yellow_tripdata_2024-09.parquet")
df.head()
```

```python
# Finding null values
df.isna().sum()
```

```python
df_clean = df.dropna(how='all')
```

```python
df_clean.shape
```

```python
df_clean['passenger_count'] = df_clean['passenger_count'].fillna(0)
```

```python
df_clean['passenger_case'] = df_clean['passenger_count'].apply(lambda x: 'Has No Passengers' if x == 0 else 'Has Passengers')
```

```python
df_clean.groupby('passenger_case')[['passenger_count']].sum()
```

```python
df_clean[df_clean['trip_distance']==0]
```

```python
df_clean['trip_distance_flag'] = df_clean['trip_distance'].apply(lambda x: 'No Distance' if x == 0 else 'Has Distance')
```

```python
df_clean.groupby('trip_distance_flag')[['trip_distance']].sum()
```

```python
df_clean.head()
```

### Filtering Out Outliers

```python
df_clean['pickup_month'] = df_clean['tpep_pickup_datetime'].dt.strftime('%Y-%m')
```

```python
df_clean.groupby('pickup_month').count()
```

```python
df_one_month = df_clean[df_clean['pickup_month'] == '2024-09']
```

```python
df_one_month.columns
```

```python
df_one_month['trip_duration'] = df_one_month['tpep_dropoff_datetime'] - df_one_month['tpep_pickup_datetime'] 
```

```python
df_one_month['trip_duration_minutes'] = df_one_month['trip_duration']/ pd.Timedelta(minutes=1) 
```

```python
df_one_month[df_one_month['trip_duration_minutes'] <= 0 ]
```

```python
df_one_month[df_one_month['trip_duration_minutes'] > 120 ]
```

```python
df_one_month.dtypes
```

```python
df_one_month['trip_duration_flag'] = df_one_month['trip_duration_minutes'].apply(lambda x: 'Invalid Trip' if x <= 0 else ('Valid Trip - Over 3hs' if x >= 120 else 'Valid Trip'))
```

```python
df_one_month.drop(columns = ['trip_duration','pickup_month'],inplace = True)
```

```python
df_one_month['fare_flag'] = df_one_month['total_amount'].apply(lambda x: 'Invalid Fare' if x < 0 else 'Valid Fare')
```

```python
df_one_month.groupby('payment_type').count()
```

```python
df_one_month.groupby(['passenger_case','trip_distance_flag','trip_duration_flag','fare_flag']).count().reset_index()
```

```python

df_final = df_one_month[(df_one_month['passenger_case'] == 'Has Passengers')
                        & (df_one_month['trip_distance_flag'] == 'Has Distance')
                        & (df_one_month['trip_duration_flag'] == 'Valid Trip')
                        & (df_one_month['fare_flag'] == 'Valid Fare')
                        & (df_one_month['payment_type'] != 3)
                        ]
```
